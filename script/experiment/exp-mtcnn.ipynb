```python
import mtcnn
import cv2
from pyprojroot.here import here
import matplotlib.pyplot as plt
#import matplotlib for image
import matplotlib.image as mpimg
```

```python
# load face on data/Anton/Anton-1.jpg and detect faces
img = cv2.imread("../data/Anton/Anton-1.jpg")
detector = mtcnn.MTCNN()
faces = detector.detect_faces(img)


```

```python
def get_face(img, box):
    x1, y1, width, height = box
    x1, y1 = abs(x1), abs(y1)
    x2, y2 = x1 + width, y1 + height
    face = img[y1:y2, x1:x2]
    return face, (x1, y1), (x2, y2)

face_crop = get_face(img, faces[0]['box'])

#plot face_crop and img side by side
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))
ax1.imshow(face_crop[0])
ax2.imshow(img)


```
